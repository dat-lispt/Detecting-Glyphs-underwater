```python
import cv2 as cv
import numpy as np
import sys
from matplotlib import pyplot as plt

raw = cv.VideoCapture("videoplayback.mp4")
if(raw.isOpened() == False):
    print("Error, Video is not open.")

while(raw.isOpened()):
    ret, frame = raw.read() 

    if(ret ==True):
        frame2 = frame.copy()
        template1= cv.imread("1st_glypth.png")
        w,h = template1.shape[::-1]
        # bw = cv.cvtColor(frame2, cv.COLOR_BGR2GRAY)
        res = cv.matchTemplate(frame2,template1,cv.TM_CCOEFF)
        min_val, max_val, min_loc, max_loc = cv.minMaxLoc(res)

        top_left = max_loc
        bottom_right = (top_left[0] + w, top_left[1]+h)
        cv.rectangle(frame2,top_left,bottom_right,255,2)

        cv.imshow("detect",res)
        cv.imshow("Show",frame2)
        cv.waitKey(200)
        if cv.waitKey(25) & 0xFF == ord('q'): 
            break
    else:
        break

raw.release()
cv.destroyAllWindows()
```
